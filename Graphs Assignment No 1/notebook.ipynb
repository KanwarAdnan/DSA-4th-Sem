```python
from numpy import arange, log
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Data
X = arange(1,100)
print(f"X = {X}")
```

```python
# Plotting Function
def myPlot(title,x,y):
  plt.title(title)
  plt.xlabel("X-Axis")
  plt.ylabel("Y-Axis")
  print(f"Y = {y}")
  plt.plot(x,y)
```

```python
# Function 1
Y = (2*X)
myPlot("Y = 2x",X,Y)
plt.show()
```

```python
# Function 2
Y = (3*X)
myPlot("Y = 3x",X,Y)
plt.show()
```

```python
# Function 3
Y = (X**3)
myPlot("Y = x^3",X,Y)
plt.show()
```

```python
# Function 4
Y = log(X)
myPlot("Y = log x",X,Y)
plt.show()
```

```python
# Function 5
Y = (2*(X**2))+(3*X)
myPlot("Y=2x^2 + 3x",X,Y)
plt.show()
```

```python
# Function 6
Y = (3*X)+2
myPlot("Y = 3x + 2",X,Y)
plt.show()
```

```python
# Function 7
Y = (X^2)
myPlot("",X,Y)
Y = 1000*X
myPlot("Y = x^2 vs Y = 1000x",X,Y)
plt.legend(['y = x^2', 'y = 1000x'])
plt.show()
```

They are intersecting at point (0,0).
